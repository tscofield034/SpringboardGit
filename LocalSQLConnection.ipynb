```python
 def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
```

```python
# Q10: Produce a list of facilities with a total revenue less than 1000.
#The output of facility name and total revenue, sorted by revenue. Remember
#that there's a different cost for guests and members!
def question_ten(conn):
    """
    """
    cur = conn.cursor()
    query = """
        SELECT * 
        FROM (
        SELECT subquery.facility, SUM( subquery.cost ) AS total_revenue
        FROM (
        SELECT f.name AS facility, 
        CASE WHEN b.memid = 0
        THEN f.guestcost * b.slots
        ELSE f.membercost * b.slots
        END AS cost
        FROM Bookings AS b
        INNER JOIN Facilities AS f
            ON b.facid = f.facid
        INNER JOIN Members AS m
            ON b.memid = m.memid
        )subquery
        GROUP BY subquery.facility
        )subquery2
        WHERE subquery2.total_revenue < 1000
        """
    cur.execute(query)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
# Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order
def question_eleven(conn):
    """
    """
    cur = conn.cursor()
    query = """
        SELECT m.surname AS member_surname, m.firstname AS member_firstname, m.recommendedby AS recommender_id,
            m2.surname AS recommender_surname, m2.firstname AS recommender_firstname
        FROM Members AS m
        LEFT JOIN Members AS m2 on m.recommendedby = m2.memid
        WHERE m.recommendedby > 0
        ORDER BY m2.surname, m2.firstname;
        """

    cur.execute(query)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
# Q12: Find the facilities with their usage by member, but not guests
def question_twelve(conn):
    """
    """
    cur = conn.cursor()
    query = """
        SELECT f.name AS facility, COUNT(b.memid) AS usage
        FROM (
        SELECT facid, memid
        FROM Bookings
        WHERE memid !=0
        ) AS b
        LEFT JOIN Facilities AS f ON b.facid = f.facid
        GROUP BY b.facid;
        """

    cur.execute(query)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
# Q13: Find the facilities usage by month, but not guests
def question_thirteen(conn):
    """
    """
    cur = conn.cursor()
    query = """
        SELECT b.months AS facilitiy_month, COUNT(b.memid) AS usage
        FROM (
        SELECT MONTH(starttime), memid
        FROM Bookings
        WHERE memid !=0
        ) AS b
        GROUP BY b.months;
        """

    cur.execute(query)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```
